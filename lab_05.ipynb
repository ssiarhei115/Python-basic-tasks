### Task: The Number Analysis Tool
* Write a program as a .py file or Jupyter notebook that analyzes a number
provided by the user.
* The program should:
    * Ask the user to choose an analysis type: "prime", "armstrong" or
"perfect".
        * A prime number is a positive integer greater than 1 that has no
positive integer divisors other than 1 and itself.
        * An Armstrong number (also known as a narcissistic number) is an
integer such that the sum of its own digits each raised to the power
of the number of digits equals the number itself.
        * A perfect number is a positive integer that is equal to the sum of its
proper positive divisors, excluding itself.
    * Implement three functions for the three analysis types. The functions
should only check an integer for the chosen property and return True or
False .
    * Implement a main function that asks the user for a number for analysis,
calls the appropriate analysis function, prints the result and asks if the
user wants to analyze another number.
    * Validate function parameters using assert statements.
    * Print whether the number meets the chosen criteria in a user-friendly way
using f-strings.
    * No variable should be defined in the global scope. All variables must be
defined within functions.
    * The program should continue to run and ask the user for new numbers
until the user decides to exit

```python
def check_prime(n):
    """ Check if the number is prime """ #2, 3, 5, 7, 11, 13, 17, 19, 23, and 29
    
    for i in range(1,n+1):
        #print(n, i)
        if n % i == 0 and all([i !=1, i != n]):
            return False
        elif i == n-1:
            return True
       
            
def check_armstrong(n):
    """ Check if the number is Armstrong """ # 153
    
    result = 0
    power = len(str(n))

    for i in str(n):
        result += int(i)**power
        
    if result == n:
        return True
    else:
        return False
    
    
def check_perfect(n):
    """ Check if the number is perfect """ # 6, 28, 496, and 8128
    
    result = 0

    for i in range(1,n):
        if n % i == 0:
            result += i
        
    if result == n:
        return True
    else:
        return False
    
    
def check_number(num, type):
    """ Choose the number check function """
    
    if type == '1':
        return check_prime(num)
    elif type == '2':
        return check_armstrong(num)
    elif type == '3':
        return check_perfect(num)
    
    
s = 153
print(s)
print('prime', check_prime(s))
print('armstrong', check_armstrong(s))
print('perfect', check_perfect(s)) 
        
```

```python
def check_number_loop():
    """ Check number in a loop """
    
    if_check = True
    result = 0

    while if_check:
        choice = input('Want to check? (y/n):')
        
        if choice != 'y':
            if_check = False
            break
        
        inp_number = input('Number to check (int):')
        inp_type = input('Check for Prime/Armstrong/Perfect? (1/2/3)?:)')
        check_types = {'1':'Prime', '2':'Armstrong', '3':'Perfect'}
        
        try:
            inp_number = int(inp_number)
        except:
            print('Number should be integer!!')
        
        assert inp_number > 0, 'Number should be > 0!'
        assert inp_type in '123', 'Check type should be 1 or 2 or 3!!'
               
        result = check_number(inp_number, inp_type)
        
        print(f"{inp_number} was checked for type {check_types[inp_type].upper()}: {result} {check_types[inp_type].upper()}")
        
        
        
check_number_loop()   
    
```
