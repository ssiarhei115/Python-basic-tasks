### Happy number

In number theory, a happy number is a number which eventually reaches 1 when the number is replaced by the sum of the square of each digit. For instance, 13 is a happy number because 

$$ 1^2 + 3^2 = 10 --> 1^{2}+0^{2}=1 $$

Ex: 7, 23, 32, 44, 49, 79, 130, 28, 82 

    Check for infinite loop.

```python
def square_num(n):
    r = 0
    for item in str(n):
        r += int(item) ** 2
    return r


def happy_num(n):
    c=1
    m=n
    while True:
        
        n = square_num(n)
        m = square_num(square_num(m))
        #print(c, n, m)
        c += 1
                
        if n == 1 or m == 1:
            return (1, c)
        if n == m:
            #print(c, n, m)
            return (0, c)
    
for i in range(3,100):
    check = happy_num(i)
    if  check[0] == 1:
        print(f"Happy number/sequence len: {i}/{check[1]}")    

```

### Number Factorial using recursion

```python
def factorial(n):
    
    print(n)
    if n == 1 or n == 0:
        return 1
    else:
        return n * factorial(n-1)
    
factorial(5)
```

## GPS time roll over

GPS broadcasts also a time, starting from January 6th, 1980

    * 10-bit week number
    * 19-bit seconds-into-week

week number counter range: 0â€“1023

integer overflow causes the counter to roll over every 1024 weeks or ~19.6 years. What are the dates of first 5 roll overs?

```python
import datetime as dt
from datetime import datetime

#st = datetime.strptime('6 Jan 1980', '%d %b %Y')
st = datetime.strptime('06/01/1980', '%d/%m/%Y')
#new_st = st.replace(hour=11, minute=59)
print(f"GPS started to broadcast time on {st}")

week_counter = 2**10
print(f"roll over every {week_counter} weeks")

ro = st
for i in range(5):
    ro += dt.timedelta(weeks=week_counter)
    print(f"{i+1} roll over: {ro}")
```

### ISBN with missing number

define a missing number in a given ISBN


```python
from sympy import symbols, solveset, solve, Eq, Mod, S
from sympy.calculus.util import continuous_domain, function_range

isbn = '9783x43910743'
x = symbols('x')
f = 9+3*7+8+3*3+x+3*4+3+3*9+1+3*0+7+3*4+3 #% 10

for q in range(10):
    if Mod(f.subs({x:q}),10) == 0:
        print(f"Missing number in ISBN {isbn}: {q}")
```
