### Task 1: The Temperature Converter
* Write a program as a .py file or Jupyter notebook that converts temperature
values between Celsius, Kelvin, and Fahrenheit.
* The program should:
    * Read the temperature value and unit (Celsius, Kelvin, or Fahrenheit) from
the user.
    * Convert the temperature to the other two units.
    * Print the results in a user-friendly way using f-strings.
    * Validate variables using assert statements, e.g. prevent negative Kelvin
temperatures.
    * Handle errors gracefully using try and except statements.
    * Ask the user if they want to convert another temperature and repeat the
process until the user chooses to exit the program by entering "exit".
* The program should be demonstrable with positive and negative
temperatures.

```python
def temp_converter(val, unit):
    """ Converts temperature values between Celsius, Kelvin, and Fahrenheit """
    
    if unit == 'C':
        t_kelv = val + 273.15
        t_fahr = val * 9/5 + 32
        return(f"{val}{unit} is equal to: \n -- tK {t_kelv} \n -- tF {t_fahr}")
    if unit == 'F':
        t_celc = (val - 32) * 5/9
        t_kelv = t_celc + 273.15
        return(f"{val}{unit} is equal to: \n -- tC {t_celc} \n -- tK {t_kelv}")
    if unit == 'K':
        t_celc = val - 273.15
        t_fahr = t_celc * 9/5 + 32 
        return(f"{val}{unit} is equal to: \n -- tC {t_celc} \n -- tF {t_fahr}")
```

```python
while True:
    
    start = input("Do you want to convert? YES | EXIT:")
    if start == 'YES':
        t_value = input('Temperature value:')
        t_unit = input('Temperature unit C|K|F:')
        
        try:
            t_value = float(t_value)
        except:
            print('Check the value!!')
            continue
        if t_unit == 'K':
            assert t_value >= 0
        assert t_unit in ['C', 'K', 'F']
        
        if t_unit == 'C':
            t_kelv = t_value + 273.15
            t_fahr = t_value * 9/5 + 32
            print(f"{t_value}{t_unit} is equal to: \n -- tK {t_kelv} \n -- tF {t_fahr}")
        elif t_unit == 'F':
            t_celc = (t_value - 32) * 5/9
            t_kelv = t_celc + 273.15
            print(f"{t_value}{t_unit} is equal to: \n -- tC {t_celc} \n -- tK {t_kelv}")
        elif t_unit == 'K':
            t_celc = t_value - 273.15
            t_fahr = t_celc * 9/5 + 32 
            print(f"{t_value}{t_unit} is equal to: \n -- tC {t_celc} \n -- tF {t_fahr}")
        
                
    else:
        break
    
```

### Task 2: The Prime Number Checker
* Write a program as a .py file or Jupyter notebook that checks if a given
number is prime.
    * A prime number is a positive integer greater than 1 that has no positive
integer divisors other than 1 and itself.
* Starting assumptions for the program:
    * An integer number is given as a variable or input by the user.
* The program should:
    * Check if the number is prime using a while loop.
        * The while loop should test divisibility with integer numbers from 2 up
to half the number.
        * If the number is divisible by any of these numbers, it is not prime.
        * If the loop completes without finding any divisors, the number is
prime.
    * Print whether the number is prime or not in a user-friendly way using f-
strings.
    * Print the smallest divisor found if the number is not prime.
    * Validate the input using assert statements to ensure it is a positive
integer greater than 1.
    * Handle errors gracefully using try and except statements.
* The program should be demonstrable with different numbers, including prime
and non-prime numbers.
* The first few prime numbers are 2, 3, 5, 7, 11, 13, 17, 19, 23, and 29.
* Hint: Use the modulus operator % to check for divisibility.
Optional deep dive: You can optimize the prime-checking algorithm by only testing
divisibility up to the square root of the number. This is because if n is a composite
number n = a × b, at least one of the factors a or b must be less than or equal to
√n.

```python
n = 9

assert type(n) is int and n > 1, 'Check the value!!'

i = 1
divisors = []

while i <= n:
    if n % i == 0:
        divisors.append(i)
    i += 1

if len(divisors) == 2:
    print(f"{n} is PRIME")
else:
    print(f"{n} is NOT prime, divisors: {divisors}")
    
```

### Task 3: The Perfect Number Finder
* Write a program as a .py file or Jupyter notebook that finds perfect numbers.
    * A perfect number is a positive integer that is equal to the sum of its
proper positive divisors, excluding itself.
* Starting assumptions for the program:
    * An lower limit is given as a variable or input by the user.
    * An upper limit is given as a variable or input by the user.
* The program should:
    * Find all perfect numbers within the given range.
        * Use a while loop to iterate through all numbers in the range.
        * Use a nested while loop to find the proper divisors of each number.
        * Sum the proper divisors and check if they equal the number.
    * Print the perfect numbers found in a user-friendly way using f-strings.
    * Print a "status" message periodically, e.g. every 100 numbers, indicating
which number is being tested right now.
    * Validate the input using assert statements to ensure the limits are
positive integers and the lower limit is less than the upper limit.
    * Handle errors gracefully using try and except statements.
* The program should be demonstrable with different ranges, including ranges
that contain no perfect numbers.
* Perfect numbers are rare. The first few are 6, 28, 496, and 8128.
* Hint: This task may run for a while, especially for larger ranges. Be patient and
debug your code with smaller ranges first.

```python
l_lim = input('Low limit:')
h_lim = input('High limit:')
#l_lim = 1
#h_lim = 1000

try:
    l_lim = int(l_lim)
    h_lim = int(h_lim)
except:
    print('check the limits!!')
  
assert all(x>0 for x in [l_lim, h_lim]), 'Limits should be > 0!!'
assert h_lim > l_lim, 'h_lim should be > l_lim'  

for number in range(l_lim, h_lim+1):
    i = 1
    divisors = []

    while i < number:
        if number % i == 0:
            divisors.append(i)
        i += 1

    if sum(divisors) == number:
        print(f"{number} is PERFECT number \n -- its proper positive divisors are: {divisors}")
```
