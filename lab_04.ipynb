### The Password Validator
* Write a program as a .py file or Jupyter notebook that validates passwords
based on specific criteria.
* The program should:
    * Read a string from the user. This is our not-real password. Don't use a
real password!
    * Validate the password based on the following criteria:
        * At least 8 characters long
        * Contains at least one uppercase letter
        * Contains at least one lowercase letter
        * Contains at least one digit
        * Contains at least one special character (e.g., !, @, #, $, %, ^, &, *)
    * Use a for loop to iterate through the characters of the password to check
for the presence of each required character type.
    * Print whether the password is valid or not in a user-friendly way using f-
strings.
    * Handle errors gracefully using try and except statements.
    * Hint: You can check for membership in a string using the in keyword,
e.g. if c in "0123456789" .

```python
%%time
s = 'u3uyyyyyyyQ!'

special = "!@#$%^&*"

cond_1 = (len(s) >=8, 'PW length should be >=8!!')
cond_2 = (any([i in special  for i in s]), 'At least one special symbol should be used!! (!, @, #, $, %, ^, &, *)')
cond_3 = (any([i.isupper() for i in s]), 'At least one uppercase letter should be used!!')
cond_4 = (any([i.islower() for i in s]), 'At least one lowerrcase letter should be used!!')
cond_5 = (any([i.isdigit() for i in s]), 'At least one digit should be used!!')

pw_ok = 0
for cond in [cond_1, cond_2, cond_3, cond_4, cond_5]:
    if not cond[0]:
        print(cond[1])
        pw_ok +=1
        
if pw_ok == 0:
    print('Password is OK!')
```

### Task 2: The Armstrong Number Checker
* Write a program as a .py file or Jupyter notebook that checks if a given
number is an Armstrong number.
    * An integer is a Armstrong number (also known as a narcissistic number) if
the sum of its own digits, each raised to the power of the number of
digits, equals the number itself.
    * For example: 153 is an Armstrong number because it has 3 digits, and 1³+ 5³ + 3³ = 153. 42 is not an Armstrong number because 4² + 2² ≠ 42.
* The program should:
    * Read an integer number from the user.
    * Test if this string is an Armstrong number using a for loop to iterate
through the digits of the number.
    * Print whether the number is an Armstrong number or not in a user-
friendly way using f-strings.
    * Validate the input using assert statements to ensure it is a positive
integer.
    * Handle errors gracefully using try and except statements.

```python
n = -153
n = 42


print(f"Number to analyse: {n}")
assert type(n) is int, 'use integer!'
assert n>0, 'use positive number!'

result = 0
power = len(str(n))

for i in str(n):
    result += int(i)**power
    print(i, result)
    
if result == n:
    print(f"Number {n} is Armstrong number")

```

### Task 3: String Compression
* Write a program as a .py file or Jupyter notebook that performs basic string
compression using the counts of repeated characters.
    * Compression is done by counting the consecutive occurrences of each
character in the string and replacing them with the character followed by
the count.
    * For example, the string "aaabcddddddaa" would be compressed to
"a3b1c1d6a2".
* The program should:
    * Read a string from the user.
    * Use a for loop to iterate through the string and build the compressed
version.
    * Print the original and compressed strings in a user-friendly way using f-
strings.
    * Print the length of both the original and compressed strings in a user-
friendly way using f-strings.
    * Validate the input using assert statements to ensure it does not
contain digits.
    * Handle errors gracefully using try and except statements.

```python
s = 'aaabc555ddddddaarrrr22'


compressed_s = '' 
pos_symbol = s[0]
pos_count = 0

for m in range(len(s)):
    
    if s[m] == pos_symbol:
        pos_count += 1
        if m == len(s) - 1:
            compressed_s += pos_symbol + '(' + str(pos_count) + ')'
    else:
        compressed_s += pos_symbol + '(' + str(pos_count) + ')'
        pos_symbol = s[m]
        pos_count = 1


print(s)
print(compressed_s) 
```
