### Task 1: The Temperature Converter
* Write a program as a .py file or Jupyter notebook that converts a temperature given in Celsius to Kelvin and Fahrenheit.
* Starting assumptions for the program: 
    * The temperature in Celsius is given as a variable or input by the user.
* The program should:
    * Convert the temperature to Kelvin and Fahrenheit.
    * Print the results in a user-friendly way using f-strings.
    * Validate variables using assert statements, e.g. prevent negative Kelvin temperatures.
* The program should be demonstrable with positive and negative temperatures.

```python
def converter(temp):
    """Temperature converter """
    
    if not any([type(temp) is float, type(temp) is int]):
        print('Check the input!!')
        return (type(temp), 'na', 'na')
    else:
        t_kelv = temp + 273.15
        t_fahr = temp * 9/5 + 32
        return (temp, t_kelv, t_fahr)

result = converter(100)

print(f"Value entered: {result[0]} \n conversion to Kelvin: {result[1]} \n conversion to Fahrengeit: {result[2]}")
```

### Task 3: The Time Duration Converter
* Write a program as a .py file or Jupyter notebook that converts a time duration given in seconds to days, hours, minutes, and seconds.
* The program should:
    * Read the time duration in seconds from the user or set it as a variable.
    * Convert the time duration to days, hours, minutes, and seconds.
* Start with the largest unit and work down to the smallest unit.
* Use an integer division operator to determine the value of each unit.
* Use the modulus operator to determine the remainder for the next smaller unit.
    * Print the results in a user-friendly way using f-strings.
* The program should be demonstrable with different time durations, including values less than a minute, less than an hour, less than a day, and more than a day.

```python
def duration(value):
    """converts a time duration given in seconds to days, hours, minutes, and seconds"""
   
    min = 60 
    hour = min * 60
    day = hour * 24
    days = hours = mins = 0

    if not any([type(value) is float, type(value) is int]) or value < 0:
        print('Check the input!!')
    else:
        days = value // day
        hours = (value - days * day) // hour
        mins = (value - days * day - hours * hour) // min
        secs = value - days * day - hours * hour - mins * min
        print(f"{value} seconds corresponds: \n {days} days, {hours} hours, {mins} mins, {secs} seconds")

duration(65)
```
